```python
grouped_x_y = micro_df[(micro_df.ScoreCancer > micro_df.ScoreCancer.quantile(0.99))].groupby(['Y','X']).mean().reset_index()
```

```python
grouped_x_y = grouped_x_y.sort_values('ScoreCancer',ascending=False)[:100]
```

```python
#merging top 1% grouped with offsite facility database to find states
m1_df = grouped_x_y.merge(offsite_df[['X','Y','State','FacilityNumber','City','Latitude','Longitude']],on=['X','Y'],how='left')
m1_df.groupby(['State']).mean().sort_values('ScoreCancer',ascending=False)
```

```python
#grouped the top 1% with facility database to find states 
m_df = grouped_x_y.merge(facility_df[['X','Y','State','FacilityNumber','City','Latitude','Longitude']],on=['X','Y'],how='left')
m_df.groupby(['State']).mean().sort_values('ScoreCancer',ascending=False)[:10]
```

```python
#facility details for PA
facility_PA = m_df.groupby('State').get_group('PA')

#offset facility details for PA
offsite_PA = m1_df.groupby('State').get_group('PA')
```

```python
def facility_details(state,facility_df,offsite_df):
    #grouped dataframe for facility
    m_df = grouped_x_y.merge(facility_df[['X','Y','State','FacilityNumber','City','Latitude','Longitude']],on=['X','Y'],how='left')
    
    #grouped dataframe for offsite
    m1_df = grouped_x_y.merge(offsite_df[['X','Y','State','FacilityNumber','City','Latitude','Longitude']],on=['X','Y'],how='left')
    
    #facility_details for state
    facility_final_df = m_df.groupby('State').get_group(state)

    #offset facility details for PA
    offsite_final_df = m1_df.groupby('State').get_group(state)
    
    return facility_final_df, offsite_final_df 
```

```python
facility_OH, offsite_OH = facility_details('OH',facility_df,offsite_df)
```

```python
facility_TX, offsite_TX = facility_details('TX',facility_df,offsite_df)
```
