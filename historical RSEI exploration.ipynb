# 2013 RSEI Aggregate analysis

```python
rsei13 = pd.read_csv(r"v234_aggregated_micro_gc14_2013.csv.gz",compression='gzip', header = None)
```

```python
rsei13
```

```python
rsei13.columns = ['X','Y','NumberOfFacilities','NumberOfReleases','NumberOfChemicals','ToxConc','Score','Pop','ScoreCancer','ScoreNonCancer']
```

```python
rsei13.head()
```

```python
grouped_by_13 = rsei13[(rsei13.ScoreCancer>rsei13.ScoreCancer.quantile(0.99))].groupby(['Y','X']).mean().reset_index()
```

```python
grouped_by_13
```

```python
rsei13_t100=grouped_by_13.sort_values('ScoreCancer',ascending=False)[:100]
```

```python
rsei13_t100
```

```python
#merging top 1% grouped with offsite facility database to find states of only top 100 ScoreCancer. 
merge_offsite_rsei13_99_df = grouped_by_13.merge(offsite[['X','Y','State','FacilityNumber','City','Latitude','Longitude']],on=['X','Y'],how='left')
merge_offsite_rsei13_99_df['year'] = 2013
merge_offsite_rsei13_99_df.groupby(['State']).mean().sort_values('ScoreCancer',ascending=False)
```

```python
# merging all of the columns of aggregated data which has top 1 percentile ScoreCancer
merge_facility_rsei13t100_df = rsei13.merge(offsite[['X','Y','State','FacilityNumber','City','Latitude','Longitude']],on=['X','Y'],how='left')
merge_facility_rsei13t100_df.groupby(['State']).mean().sort_values('ScoreCancer', ascending = False)
```

```python
#merging top 1% grouped with facility database to find states of only top 100 ScoreCancer. 
merge_facility_rsei13_99_df = grouped_by_13.merge(facility[['X','Y','State','FacilityNumber','City','Latitude','Longitude']],on=['X','Y'],how='left')
merge_facility_rsei13_99_df.groupby(['State']).mean().sort_values('ScoreCancer',ascending=False)
```

```python
def facility_details_13(state,facility_df,offsite_df):
    #grouped dataframe for facility
    m_df = grouped_by_13.merge(facility_df[['X','Y','State','FacilityNumber','City','Latitude','Longitude']],on=['X','Y'],how='left')
    
    #grouped dataframe for offsite
    m1_df = grouped_by_13.merge(offsite_df[['X','Y','State','FacilityNumber','City','Latitude','Longitude']],on=['X','Y'],how='left')
    
    #facility_details for state
    facility_final_df = m_df.groupby('State').get_group(state)

    #offset facility details for PA
    offsite_final_df = m1_df.groupby('State').get_group(state)
    
    return facility_final_df, offsite_final_df 
```

```python
#retriving values of PA, TX, OH which were top 100 contributers to ScoreCancer
facility_OH_13, offsite_OH_13 = facility_details_13('OH',facility,offsite)
```

```python
facility_OH_13
```

```python
facility_PA_13, offsite_PA_13 = facility_details_13('PA',facility,offsite)
```

```python
facility_TX_13, offsite_TX_13 = facility_details_13('TX',facility,offsite)
```
